```python
a = np.array([[5,2,9,4],[1,8,8,9],[6,6,4,7],[3,5,6,2]])
a
# Seleccionar columna 2
a[:,2]
# Seleccionar fila 2
a[2,:]
# Selcción de un subarreglo
a[0:2,1:3]
# Selección condicional
np.where(a%2==0,a,0)
a>5
a*3
# Determinante de un arreglo
np.linalg.det(a)
b = np.array([[5, 4, 2, 0], [9, 9, 6, 1], [3, 0, 5, 3], [8, 0, 6, 1]])
b
c = a * b
c
```
