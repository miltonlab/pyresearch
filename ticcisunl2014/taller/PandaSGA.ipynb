```python
dfpregrado=pregrado.copy()
```

```python
dfpregrado=dfpregrado.loc[:, ['id', 'nombre', 'matriculados']]
```

```python
dfpregrado=dfpregrado.sort(columns=['id'])
```

```python
dfpregrado['aprobados'] = 0
```

```python
dfpregrado['reprobados'] = 0
```

```python
for index, row in dfpregrado.iterrows():
    id_oferta = row['id']
    wsmat=urllib.urlopen('http://'+uws+':'+pws+'@ws.unl.edu.ec/sgaws/wsestadistica/sgaws_nmatriculados_oferta?id_oferta=%s' % id_oferta)
    wsapr=urllib.urlopen('http://'+uws+':'+pws+'@ws.unl.edu.ec/sgaws/wsestadistica/sgaws_naprobados_oferta?id_oferta=%s' % id_oferta)
    wsrepr=urllib.urlopen('http://'+uws+':'+pws+'@ws.unl.edu.ec/sgaws/wsestadistica/sgaws_nreprobados_oferta?id_oferta=%s' % id_oferta)
    jsmat=pd.read_json(wsmat.read()[8:-9])
    jsapr=pd.read_json(wsapr.read()[8:-9])
    jsrepr=pd.read_json(wsrepr.read()[8:-9])
    #print strjs
    mats=jsmat.iloc[-1,0]
    aprs=jsapr.iloc[-1,0] 
    reprs=jsrepr.iloc[-1,0] 
    dfpregrado.loc[dfpregrado.id == id_oferta, ['matriculados']] = mats
    dfpregrado.loc[dfpregrado.id == id_oferta, ['aprobados']] = aprs
    dfpregrado.loc[dfpregrado.id == id_oferta, ['reprobados']] = reprs
    #row['matriculados'] = mats
    wsmat.close()
    wsapr.close()
    wsrepr.close()
```

```python
#dfcis = pd.DataFrame(columns=['oferta','matriculados'])
data = []
import urllib
urlwsstat = 'http://%s:%s@ws.unl.edu.ec/sgaws/wsestadistica' % (uws, pws)
id_carrera = 79 # Sistemas
for id_oferta, nombre in ids_ofertas.values:
    wsmat=urllib.urlopen('%s/sgaws_nmatriculados_carrera?id_oferta=%s&id_carrera=%s' % (urlwsstat, id_oferta, id_carrera))
    jsmat=pd.read_json(wsmat.read()[8:-9])
    mats=jsmat.iloc[-1,0]
    data.append({'oferta':id_oferta, 'nombre':nombre, 'matriculados':mats})
    wsmat.close()
dfcis = pd.DataFrame(data) 
```
