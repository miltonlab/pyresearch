```python
z = randn(100,2)
z[:,1] = 0.5*z[:,0] + numpy.sqrt(0.5)*z[:,1]
x = z[:,0]
y = z[:,1]
plt.subplot(1,3,1)
plt.scatter(x, y, label='Scatter 1')
#
plt.subplot(1,3,2)
plt.scatter(x, y, s=60, c='#FF7F00', marker='o', alpha=.5, label='Scatter 2')
#
size_data = numpy.exp(numpy.exp(numpy.exp(rand(100))))
size_data = 200 * size_data/numpy.amax(size_data)
size_data[size_data<1]=1.0
plt.subplot(1,3,3)
plt.scatter(x,y, s = size_data, c = '#FF7F00', marker='s', label = 'Scatter 3')
```
